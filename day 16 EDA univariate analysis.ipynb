```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head(3)
```

# Working With Categorical Data

# 1 Countplot

```python
sns.countplot(x=df['Survived'],data=df)
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x=df['Pclass'],data=df)
```

```python
df['Pclass'].value_counts().plot(kind='bar')
```

# Pie Chart

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%2.1f')
```

# Numerical Data

# Histogram

```python
plt.hist(x=df['Age'],bins=20)
```

```python
plt.hist(df['Fare'])
```

# Distplot

```python
sns.distplot(x=df['Fare'],kde=False)
```

```python
sns.distplot(df['Age'])
```

```python
sns.boxenplot(df['Age'])
```
