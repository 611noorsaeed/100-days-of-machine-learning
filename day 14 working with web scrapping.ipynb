```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.businesslist.pk/category/money-changers/city:faisalabad").text
```

```python
soup = BeautifulSoup(page,'lxml')
```

```python
name = []
for i in soup.find_all('h4'):
    name.append(i.text)
```

```python
address = []

for i in soup.find_all('div',class_='address'):
    address.append(i.text)
```

```python
con = []

for i in soup.find_all('div',class_='s'):
    con.append(i.text)
```

```python
phone = []
email = []
for i in con:
    if i == 'Phone':
        phone.append(i)
    if i == 'E-mail':
        email.append(i)
```

```python
email
```

```python
dic = {'name':name,'address':address,'phone':phone,'email':email}
```

```python
df = pd.DataFrame(dic)
```

```python
df
```

```python
address = []
for i in soup.find_all('div',class_='address'):
    address.append(i.text)
```

```python
cont = []
for i in soup.find_all('div',class_='s'):
    cont.append(i.text)
```

```python
print(len(name))
print(len(cont))
print(len(address))
```

```python
phone = []
email = []

for i in cont:
    if i == 'Phone':
        phone.append(i)
    if i == 'E-mail':
        email.append(i)
```
